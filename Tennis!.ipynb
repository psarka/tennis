```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
import pickle

from scipy.stats import beta
from pystan import stan, StanModel
from hashlib import md5

from rpy2.robjects.packages import importr
import rpy2.robjects as ro
import pandas.rpy.common as com


%matplotlib inline

def stan_cache(model_code, model_name=None, **kwargs):
    """Use just as you would `stan`"""
    code_hash = md5(model_code.encode('ascii')).hexdigest()
    if model_name is None:
        cache_fn = 'cached-model-{}.pkl'.format(code_hash)
    else:
        cache_fn = 'cached-{}-{}.pkl'.format(model_name, code_hash)
    try:
        sm = pickle.load(open(cache_fn, 'rb'))
    except:
        sm = StanModel(model_code=model_code)
        with open(cache_fn, 'wb') as f:
            pickle.dump(sm, f)
    else:
        print("Using cached StanModel")
    return sm.sampling(**kwargs)
```

```python
df = pd.read_csv('Wimbledon_2013_Sample_Point_by_Point_data.csv')
df = df[df.Event == 'MS']
df = df.drop(['Tournament', 'Year', 'Court', 'Event', 'Round', '1st Serve Timestamp', '2nd Serve Timestamp', 'End of Point Timestamp'], axis=1)
df.rename(columns={'Tiebreak?': 'Tiebreak',
                   'Service (0=Double Fault|1=First Serve|2=Second Serve)': 'Service', 
                   'Speed MPH (where avail)': 'Speed', 
                   'Shot Main (1=FH|2=BH|3=ACE|4=Service Winner|5=Double Fault|6=Other)': 'Main_shot',
                   'Shot Detail (0=None|1=Drive|2=Return|3=Passing|4=Volley|5=Smash|6=Stroke at Net|7=Passing-Volley|8=Drop-Shot|9=Lob|10=Approach|11=Approach-Return)': 'Shot_detail',
                   'Shot Win/Err (0=Undefined|1=Winner|2=Unforced Error|3=Forced Error)': 'Shot_win/err'
                   },inplace=True)
df.Server = df.Server.apply(lambda x: np.int(x[0]))
```

```python
def f(m):
    s_0, s_1, s_2 = sum(m.Service == 0), sum(m.Service == 1), sum(m.Service == 2)
    
    # Times served 1st serve:
    n_try1 = s_0 + s_1 + s_2
    
    # Times won 1st serve:
    n_win1 = sum((m.Service == 1) & (m.Server == m.Winner))
    
    # MLE probability to win 1st serve:
    p_mle_win1 = n_win1/n_try1
    
    # Times served 2nd serve:
    n_try2 = s_0 + s_2
    
    # Times won 2nd serve:
    n_win2 = sum((m.Service == 2) & (m.Server == m.Winner))
    
    # MLE probability to win 2nd serve
    p_mle_win2 = n_win2/n_try2
        
    return pd.Series({'n_try1': n_try1, 
                      'n_try2': n_try2,
                      'n_win1': n_win1, 
                      'n_win2': n_win2,
                      'p_mle_win1': p_mle_win1,
                      'p_mle_win2': p_mle_win2,
                      })

df = df.groupby(['Match', 'Player 1', 'Player 2', 'Server']).apply(f)
df['p_mle_delta'] = df.p_mle_win2 - df.p_mle_win1
df = df.sort('p_mle_delta')
df.head()
```

```python
ro.r('library(PropCIs)')
rdf = com.convert_to_r_dataframe(df[['n_win1', 'n_try1', 'n_win2', 'n_try2']])
ro.globalenv['data'] = rdf
cis = ro.r('mapply(diffscoreci, data$n_win2, data$n_try2, data$n_win1, data$n_try1, .95)')
```

```python
df['2.5%_freq'] = [i[0] for i in cis]
df['97.5%_freq'] = [i[1] for i in cis]
```

```python
df.head()
```

```python
def f(x):
    return list(x.values.astype(np.int))

stan_data = {'N': len(df), 
             'try1': f(df.n_try1), 'try2': f(df.n_try2),
             'win1': f(df.n_win1), 'win2': f(df.n_win2)}

stan_code = '''
data {
    int<lower=1> N;
    int try1[N];
    int try2[N];
    int win1[N];
    int win2[N];
}
parameters {
    real<lower=0,upper=1> phi1;
    real<lower=0,upper=1> phi2;
    real<lower=0.1> lambda1;
    real<lower=0.1> lambda2;
    vector<lower=0,upper=1>[N] theta1;
    vector<lower=0,upper=1>[N] theta2;
}
transformed parameters {
    real<lower=0> alpha1;
    real<lower=0> alpha2;
    real<lower=0> beta1;
    real<lower=0> beta2;
    alpha1 <- lambda1*phi1;
    alpha2 <- lambda2*phi2;
    beta1 <- lambda1*(1-phi1);
    beta2 <- lambda2*(1-phi2);
}
model {
    phi1 ~ beta(1,1);
    phi2 ~ beta(1,1);
    lambda1 ~ pareto(0.1, 1.5);
    lambda2 ~ pareto(0.1, 1.5);
    
    for (i in 1:N) {
        theta1[i] ~ beta(alpha1, beta1);
        theta2[i] ~ beta(alpha2, beta2);
        win1[i] ~ binomial(try1[i], theta1[i]);
        win2[i] ~ binomial(try2[i], theta2[i]);
    }
}
generated quantities {
    vector[N] delta;
    for (i in 1:N)
        delta[i] <- theta2[i] - theta1[i];
}
'''
```

```python
fit = stan_cache(model_code=stan_code, data=stan_data)
summary = fit.summary()
```

```python
def split(namesids):
    names = [i.split('[')[0] for i in namesids]
    ids = [int(i.split('[')[1][:-1]) if '[' in i else 0 for i in namesids]
    return names, ids

res = pd.DataFrame(data=summary['summary'], columns=summary['summary_colnames'])
names, ids = split(summary['summary_rownames'])
res['name'] = names
res['id'] = ids
res = res.set_index(['name', 'id'])
res.tail(10)
```

```python
z = res.loc['delta', :][['2.5%', 'mean', '97.5%']]
z.head()
```

```python
ntop = 20
nbot = 20
top = list(range(nbot))
bot = list(range(len(z)-ntop, len(z)))

ids = top + bot

fig,(ax1,ax2) = plt.subplots(2,1,sharex=True, figsize=(8, 12))

ax1.hlines(ids[::-1], z['2.5%'].iloc[ids], z['97.5%'].iloc[ids], color = 'red', lw=2, alpha= .5)
ax1.plot(z['mean'].values[ids], ids[::-1], 'r^', alpha=.5)
ax1.hlines(ids[::-1], df['2.5%_freq'].iloc[ids], df['97.5%_freq'].iloc[ids], color = 'green', lw=2, alpha=.5)
ax1.plot(df['p_mle_delta'].values[ids], ids[::-1], 'g^', alpha=.5)

ax2.hlines(ids[::-1], z['2.5%'].iloc[ids], z['97.5%'].iloc[ids], color = 'red', lw=2, alpha= .5)
ax2.plot(z['mean'].values[ids], ids[::-1], 'r^', alpha=.5)
ax2.hlines(ids[::-1], df['2.5%_freq'].iloc[ids], df['97.5%_freq'].iloc[ids], color = 'green', lw=2, alpha=.5)
ax2.plot(df['p_mle_delta'].values[ids], ids[::-1], 'g^', alpha=.5)

ax1.set_ylim(len(z)-ntop - 2, len(z))
ax1.vlines(0, len(z)-ntop - 2, len(z), linestyles='--')
ax2.set_ylim(-1, nbot + 1)
ax2.vlines(0, -1, nbot + 1, linestyles='--')


plt.xlim(-1, 1)

ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.xaxis.tick_top()
ax1.tick_params(labeltop='off') # don't put tick labels at the top
ax2.xaxis.tick_bottom()

d = .05 # how big to make the diagonal lines in axes coordinates
# arguments to pass plot, just so we don't keep repeating them
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d,+d),(-d,+d), **kwargs)      # top-left diagonal
ax1.plot((1/2-d,1/2+d),(-d,+d), **kwargs)
ax1.plot((1-d,1+d),(-d,+d), **kwargs)    # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d,+d),(1-d,1+d), **kwargs)   # bottom-left diagonal
ax2.plot((1/2-d,1/2+d),(1-d,1+d), **kwargs)
ax2.plot((1-d,1+d),(1-d,1+d), **kwargs) # bottom-right diagonal

pairs = [(quat[quat[3]], quat[3 - quat[3]]) for quat in df.index]
labels = ["%s (vs. %s)" % pair for pair in pairs][::-1]
tlab = [labels[i] for i in top]
blab = [labels[i] for i in bot]

ax1.set_yticks(bot)
ax1.set_yticklabels(blab)

ax2.set_yticks(top)
ax2.set_yticklabels(tlab)

plt.show()
```
